```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Subir imagen
uploaded = files.upload()

# Obtener nombre automáticamente
nombre_archivo = list(uploaded.keys())[0]

# Cargar en escala de grises
img = cv2.imread(nombre_archivo, 0)

# Verificar
if img is None:
    print("Error: no se cargó la imagen")
else:
    filas, columnas = img.shape

    # Centro de la imagen
    centro = (columnas//2, filas//2)

    # Matriz de rotación
    M = cv2.getRotationMatrix2D(centro, 45, 1)

    # Aplicar rotación
    rotada = cv2.warpAffine(img, M, (columnas, filas))
```

```python
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Imagen original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(rotada, cmap='gray')
plt.title("Rotación 45°")
plt.axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Subir imagen
uploaded = files.upload()
nombre_archivo = list(uploaded.keys())[0]

# Cargar en grises
img = cv2.imread(nombre_archivo, 0)

# -------------------------
# REDUCIR CONTRASTE
# -------------------------
# alpha < 1 reduce contraste
bajo_contraste = cv2.convertScaleAbs(img, alpha=0.5, beta=50)

# -------------------------
# MOSTRAR RESULTADOS
# -------------------------
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(bajo_contraste, cmap='gray')
plt.title("Bajo contraste")
plt.axis("off")

plt.show()
```

```python
plt.hist(bajo_contraste.ravel(), 256, [0,256])
plt.title("Histograma - Bajo contraste")
plt.show()
```

```python
ecualizada = cv2.equalizeHist(bajo_contraste)
```

```python
plt.figure(figsize=(12,6))

# Imágenes
plt.subplot(2,2,1)
plt.imshow(bajo_contraste, cmap='gray')
plt.title("Bajo contraste")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(ecualizada, cmap='gray')
plt.title("Ecualizada")
plt.axis("off")

# Histogramas
plt.subplot(2,2,3)
plt.hist(bajo_contraste.ravel(), 256, [0,256])
plt.title("Histograma bajo contraste")

plt.subplot(2,2,4)
plt.hist(ecualizada.ravel(), 256, [0,256])
plt.title("Histograma ecualizado")

plt.show()
```
